```python
# Загрузите модуль pyplot библиотеки matplotlib с псевдонимом plt, а также библиотеку numpy с псевдонимом np.
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Примените магическую функцию %matplotlib inline для отображения графиков в Jupyter Notebook 
# и настройки конфигурации ноутбука со значением 'svg' для более четкого отображения графиков.
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
# Создайте список под названием x с числами 1, 2, 3, 4, 5, 6, 7 и список y с числами 3.5, 3.8, 4.2, 4.5, 5, 5.5, 7.
x = [1, 2, 3, 4, 5, 6, 7]
y = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]
print("x =", x, "и y =", y)
```

```python
# С помощью функции plot постройте график, соединяющий линиями точки с горизонтальными координатами из списка x и вертикальными - из списка y.
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Линейный график')
plt.show()
```

```python
# Затем в следующей ячейке постройте диаграмму рассеяния (другие названия - диаграмма разброса, scatter plot).
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Диаграмма рассеяния')
plt.show()
```

### Задание 2

```python
# С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно.
t = np.linspace(0, 10, 51)
print(t)
```

```python
# Создайте массив Numpy под названием f, содержащий косинусы элементов массива t.
f = np.cos(t)
print(f)
```

```python
# Постройте линейную диаграмму, используя массив t для координат по горизонтали, а массив f - для координат по вертикали. 
# Линия графика должна быть зеленого цвета.
plt.plot(t, f, color='green')

# Выведите название диаграммы - 'График f(t)'. Также добавьте названия для горизонтальной оси - 'Значения t' и для вертикальной - 'Значения f'.
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')

# Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.
plt.xlim(0.5, 9.5)
plt.ylim(-2.5, 2.5)

# Отображение графика
plt.show()
```

### *Задание 3

```python
# С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.
x = np.linspace(-3, 3, 51)
print(x)
```

```python
# Создайте массивы y1, y2, y3, y4 по следующим формулам:
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
print(y1, "\n\n", y2, "\n\n", y3, "\n\n", y4)
```

```python
# Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig и массив объектов Axes под названием ax, 
fig, ax = plt.subplots(2, 2, figsize=(8, 6))

# причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов. 
# В каждом графике массив x используется для координат по горизонтали. В левом верхнем графике для координат по вертикали используйте y1, 
# в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4. Дайте название графикам: 'График y1', 'График y2' и т.д.
# Для графика в левом верхнем углу установите границы по оси x от -5 до 5. 
# Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали. 
# Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.

# Построение графиков
ax[0, 0].plot(x, y1)
ax[0, 0].set_title('График y1')
ax[0, 0].set_xlim(-5, 5)

ax[0, 1].plot(x, y2)
ax[0, 1].set_title('График y2')

ax[1, 0].plot(x, y3)
ax[1, 0].set_title('График y3')

ax[1, 1].plot(x, y4)
ax[1, 1].set_title('График y4')

# Вертикальные и горизонтальные зазоры между графиками
plt.subplots_adjust(wspace=0.3, hspace=0.3)

# Отображение графиков
plt.show()
```

##### Импортируйте библиотеку Pandas, а также используйте для графиков стиль “fivethirtyeight”.

```python
import pandas as pd

# Установим стиль для графиков
plt.style.use('fivethirtyeight')
```

```python
# Загружаем данные
df = pd.read_csv('creditcard.csv')
df.head()
```

##### Посчитайте с помощью метода value_counts количество наблюдений для каждого значения целевой переменной Class и примените к полученным данным метод plot, чтобы построить столбчатую диаграмму.

```python
# Количество наблюдений каждого класса
class_counts = df['Class'].value_counts()

# Построим столбчатую диаграмму
class_counts.plot(kind='bar')

plt.xlabel('Class')
plt.ylabel('Количество наблюдений')
plt.show()
```

##### Затем постройте такую же диаграмму, используя логарифмический масштаб.

```python
# Построим столбчатую диаграмму с логарифмическим масштабом
class_counts.plot(kind='barh', log=True)
plt.title('Количество наблюдений каждого класса (логарифмический масштаб)')

plt.show()
```

##### На следующем графике постройте две гистограммы по значениям признака V1 - одну для мошеннических транзакций (Class равен 1) и другую - для обычных (Class равен 0). Подберите значение аргумента density так, чтобы по вертикали графика было расположено не число наблюдений, а плотность распределения. Число бинов должно равняться 20 для обеих гистограмм, а коэффициент alpha сделайте равным 0.5, чтобы гистограммы были полупрозрачными и не загораживали друг друга. Создайте легенду с двумя значениями: “Class 0” и “Class 1”. Гистограмма обычных транзакций должна быть серого цвета, а мошеннических - красного. Горизонтальной оси дайте название “V1”.

```python
# Выберем мошеннические и обычные транзакции
fraud_transactions = df[df['Class'] == 1]['V1']
normal_transactions = df[df['Class'] == 0]['V1']

# Построим две гистограммы
plt.hist(normal_transactions, bins=20, density=True, alpha=0.5, color='grey', label='Class 0')
plt.hist(fraud_transactions, bins=20, density=True, alpha=0.5, color='red', label='Class 1')

# Настроим легенду и подписи
plt.legend(['Class 0', 'Class 1'])
plt.title('Плотность распределение признака V1 для различных классов')
plt.show()
```
